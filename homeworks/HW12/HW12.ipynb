```python
import json
import random
from dataclasses import asdict, dataclass
from pathlib import Path

import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import torch
from sklearn.linear_model import Ridge
from sklearn.metrics import mean_absolute_error, mean_squared_error
from sklearn.preprocessing import StandardScaler
from torch import nn
from torch.utils.data import DataLoader, Dataset

plt.rcParams["figure.figsize"] = (14, 5)
plt.rcParams["axes.grid"] = True
pd.set_option("display.max_columns", 100)
pd.set_option("display.width", 140)

ROOT = Path.cwd()
DATA_PATH = ROOT / "S12-hw-dataset.csv"
ARTIFACTS_DIR = ROOT / "artifacts"
FIGURES_DIR = ARTIFACTS_DIR / "figures"
RUNS_PATH = ARTIFACTS_DIR / "runs.csv"
BEST_GRU_WEIGHTS_PATH = ARTIFACTS_DIR / "best_gru.pt"
BEST_GRU_CONFIG_PATH = ARTIFACTS_DIR / "best_gru_config.json"

ARTIFACTS_DIR.mkdir(exist_ok=True)
FIGURES_DIR.mkdir(exist_ok=True)

# Базовые настройки воспроизводимости и устройства.
def set_seed(seed: int = 42) -> None:
    random.seed(seed)
    np.random.seed(seed)
    torch.manual_seed(seed)
    if torch.cuda.is_available():
        torch.cuda.manual_seed_all(seed)


SEED = 42
set_seed(SEED)
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
print("Device:", device)

# Метрики.
def rmse(y_true, y_pred):
    return float(np.sqrt(mean_squared_error(y_true, y_pred)))


def mape(y_true, y_pred):
    y_true = np.asarray(y_true)
    y_pred = np.asarray(y_pred)
    mask = np.abs(y_true) > 1e-8
    return float(np.mean(np.abs((y_true[mask] - y_pred[mask]) / y_true[mask])) * 100)


def collect_metrics(y_true, y_pred):
    return {
        "mae": float(mean_absolute_error(y_true, y_pred)),
        "rmse": rmse(y_true, y_pred),
        "mape": mape(y_true, y_pred),
    }


def make_run_record(
    experiment_id,
    model_summary,
    features_summary,
    window_size,
    scaler,
    optimizer,
    lr,
    epochs_trained,
    val_metrics,
    split_summary,
    test_metrics=None,
    notes="",
):
    test_metrics = test_metrics or {}
    return {
        "experiment_id": experiment_id,
        "task": "forecasting",
        "dataset": DATA_PATH.name,
        "seed": SEED,
        "split_summary": json.dumps(split_summary, ensure_ascii=False),
        "window_size": window_size,
        "horizon": 1,
        "model_summary": model_summary,
        "features_summary": features_summary,
        "scaler": scaler,
        "optimizer": optimizer,
        "lr": lr,
        "epochs_trained": epochs_trained,
        "best_val_mae": val_metrics["mae"],
        "best_val_rmse": val_metrics["rmse"],
        "best_val_mape": val_metrics["mape"],
        "test_mae": test_metrics.get("mae"),
        "test_rmse": test_metrics.get("rmse"),
        "test_mape": test_metrics.get("mape"),
        "notes": notes,
    }


# Загрузка и sanity-check данных.
df = pd.read_csv(DATA_PATH)
df["date"] = pd.to_datetime(df["date"])
df = df.sort_values("date").reset_index(drop=True)
df["target"] = pd.to_numeric(df["target"], errors="coerce")

display(df.head())
print("Shape:", df.shape)
print("Date range:", df["date"].min(), "->", df["date"].max())
print("Missing values:")
display(df.isna().sum().to_frame("missing_count").T)
print("Most common frequency:", df["date"].diff().mode().iloc[0])

fig, ax = plt.subplots(figsize=(15, 5))
ax.plot(df["date"], df["target"], color="tab:blue", linewidth=1.2)
ax.set_title("Target over time")
ax.set_xlabel("Date")
ax.set_ylabel("target")
plt.tight_layout()
plt.show()

start_mean = df["target"].iloc[: 24 * 7].mean()
end_mean = df["target"].iloc[-24 * 7 :].mean()
print("Краткий комментарий:")
print("1. Ряд почасовой, без пропусков, с заметной изменчивостью.")
print("2. Средний уровень в конце периода отличается от начала, значит ряд не идеально стационарен.")
print("3. Для такой задачи нельзя использовать random split, иначе появится утечка будущего.")

# Temporal split без перемешивания.
n_total = len(df)
train_end = int(n_total * 0.70)
val_end = int(n_total * 0.85)

train_df = df.iloc[:train_end].copy()
val_df = df.iloc[train_end:val_end].copy()
test_df = df.iloc[val_end:].copy()

split_summary = {
    "train_rows": len(train_df),
    "val_rows": len(val_df),
    "test_rows": len(test_df),
    "train_range": [str(train_df["date"].min()), str(train_df["date"].max())],
    "val_range": [str(val_df["date"].min()), str(val_df["date"].max())],
    "test_range": [str(test_df["date"].min()), str(test_df["date"].max())],
}
print(json.dumps(split_summary, indent=2, ensure_ascii=False))

fig, ax = plt.subplots(figsize=(15, 5))
ax.plot(train_df["date"], train_df["target"], label="train", color="tab:blue")
ax.plot(val_df["date"], val_df["target"], label="validation", color="tab:orange")
ax.plot(test_df["date"], test_df["target"], label="test", color="tab:green")
ax.axvline(train_df["date"].max(), color="black", linestyle="--", alpha=0.7)
ax.axvline(val_df["date"].max(), color="black", linestyle="--", alpha=0.7)
ax.set_title("Temporal split")
ax.legend()
plt.tight_layout()
plt.savefig(FIGURES_DIR / "series_split.png", dpi=150, bbox_inches="tight")
plt.show()

# Признаки только из прошлого.
def add_time_features(source_df: pd.DataFrame) -> pd.DataFrame:
    out = source_df.copy()
    out["lag_1"] = out["target"].shift(1)
    out["lag_7"] = out["target"].shift(7)
    out["lag_14"] = out["target"].shift(14)
    out["lag_24"] = out["target"].shift(24)
    out["lag_168"] = out["target"].shift(168)
    out["rolling_mean_7"] = out["target"].shift(1).rolling(7).mean()
    out["rolling_std_7"] = out["target"].shift(1).rolling(7).std()
    out["rolling_mean_24"] = out["target"].shift(1).rolling(24).mean()
    out["rolling_std_24"] = out["target"].shift(1).rolling(24).std()
    out["hour"] = out["date"].dt.hour
    out["dayofweek"] = out["date"].dt.dayofweek
    out["dayofmonth"] = out["date"].dt.day
    out["month"] = out["date"].dt.month
    out["is_weekend"] = (out["dayofweek"] >= 5).astype(int)
    return out


feature_df = add_time_features(df).dropna().reset_index(drop=True)
feature_columns = [
    "lag_1",
    "lag_7",
    "lag_14",
    "lag_24",
    "lag_168",
    "rolling_mean_7",
    "rolling_std_7",
    "rolling_mean_24",
    "rolling_std_24",
    "hour",
    "dayofweek",
    "dayofmonth",
    "month",
    "is_weekend",
]
display(feature_df.head())

effective_train_end = feature_df.index[feature_df["date"] <= train_df["date"].max()].max() + 1
effective_val_end = feature_df.index[feature_df["date"] <= val_df["date"].max()].max() + 1

baseline_train_df = feature_df.iloc[:effective_train_end].copy()
baseline_val_df = feature_df.iloc[effective_train_end:effective_val_end].copy()
baseline_test_df = feature_df.iloc[effective_val_end:].copy()

# Обязательные baseline-эксперименты.
baseline_runs = []
y_val = baseline_val_df["target"].to_numpy()

b1_val_pred = baseline_val_df["lag_1"].to_numpy()
b1_val_metrics = collect_metrics(y_val, b1_val_pred)
baseline_runs.append(
    make_run_record(
        "B1",
        "naive last observation",
        "lag_1",
        1,
        "none",
        "none",
        None,
        0,
        b1_val_metrics,
        split_summary,
        notes="Validation only at model-selection stage",
    )
)

b2_val_pred = baseline_val_df["rolling_mean_24"].to_numpy()
b2_val_metrics = collect_metrics(y_val, b2_val_pred)
baseline_runs.append(
    make_run_record(
        "B2",
        "moving average, window=24",
        "rolling_mean_24 based on past values only",
        24,
        "none",
        "none",
        None,
        0,
        b2_val_metrics,
        split_summary,
        notes="Validation only at model-selection stage",
    )
)

x_scaler = StandardScaler()
X_train = x_scaler.fit_transform(baseline_train_df[feature_columns])
X_val = x_scaler.transform(baseline_val_df[feature_columns])
y_train = baseline_train_df["target"].to_numpy()
ridge = Ridge(alpha=1.0)
ridge.fit(X_train, y_train)
b3_val_pred = ridge.predict(X_val)
b3_val_metrics = collect_metrics(y_val, b3_val_pred)
baseline_runs.append(
    make_run_record(
        "B3",
        "Ridge(alpha=1.0)",
        ", ".join(feature_columns),
        168,
        "StandardScaler(train only)",
        "closed form / sklearn",
        None,
        1,
        b3_val_metrics,
        split_summary,
        notes="Uses lag, rolling and calendar features without future leakage",
    )
)

baseline_results_df = pd.DataFrame(baseline_runs)
display(
    baseline_results_df[
        ["experiment_id", "best_val_mae", "best_val_rmse", "best_val_mape", "model_summary"]
    ].sort_values("best_val_mae")
)

# Оконное представление для GRU.
@dataclass
class GRUConfig:
    window_size: int = 48
    batch_size: int = 64
    hidden_size: int = 64
    num_layers: int = 1
    dropout: float = 0.0
    lr: float = 1e-3
    epochs: int = 30
    patience: int = 8


class SequenceSliceDataset(Dataset):
    def __init__(self, scaled_series: np.ndarray, start_target_idx: int, end_target_idx: int, window_size: int):
        self.series = scaled_series.astype(np.float32)
        self.start_target_idx = start_target_idx
        self.end_target_idx = end_target_idx
        self.window_size = window_size

    def __len__(self):
        return max(0, self.end_target_idx - self.start_target_idx)

    def __getitem__(self, idx):
        target_idx = self.start_target_idx + idx
        x = self.series[target_idx - self.window_size : target_idx]
        y = self.series[target_idx]
        return torch.tensor(x[:, None]), torch.tensor([y])


class GRURegressor(nn.Module):
    def __init__(self, input_size=1, hidden_size=64, num_layers=1, dropout=0.0):
        super().__init__()
        self.gru = nn.GRU(
            input_size=input_size,
            hidden_size=hidden_size,
            num_layers=num_layers,
            dropout=dropout if num_layers > 1 else 0.0,
            batch_first=True,
        )
        self.head = nn.Linear(hidden_size, 1)

    def forward(self, x):
        out, _ = self.gru(x)
        return self.head(out[:, -1, :])


gru_config = GRUConfig()
full_target = df["target"].to_numpy()
target_scaler = StandardScaler()
target_scaler.fit(full_target[:train_end].reshape(-1, 1))
scaled_full_target = target_scaler.transform(full_target.reshape(-1, 1)).reshape(-1)

train_ds = SequenceSliceDataset(scaled_full_target, gru_config.window_size, train_end, gru_config.window_size)
val_ds = SequenceSliceDataset(scaled_full_target, train_end, val_end, gru_config.window_size)
test_ds = SequenceSliceDataset(scaled_full_target, val_end, len(df), gru_config.window_size)

train_loader = DataLoader(train_ds, batch_size=gru_config.batch_size, shuffle=False)
val_loader = DataLoader(val_ds, batch_size=gru_config.batch_size, shuffle=False)
test_loader = DataLoader(test_ds, batch_size=gru_config.batch_size, shuffle=False)


def inverse_transform_target(values):
    values = np.asarray(values).reshape(-1, 1)
    return target_scaler.inverse_transform(values).reshape(-1)


def evaluate_model(model, loader):
    model.eval()
    preds = []
    trues = []
    with torch.no_grad():
        for xb, yb in loader:
            xb = xb.to(device)
            yb = yb.to(device)
            outputs = model(xb)
            preds.append(outputs.squeeze(-1).cpu().numpy())
            trues.append(yb.squeeze(-1).cpu().numpy())

    preds = np.concatenate(preds)
    trues = np.concatenate(trues)
    preds_inv = inverse_transform_target(preds)
    trues_inv = inverse_transform_target(trues)
    return collect_metrics(trues_inv, preds_inv), trues_inv, preds_inv


# Обучение GRU с выбором лучшей эпохи по validation MAE.
model = GRURegressor(
    hidden_size=gru_config.hidden_size,
    num_layers=gru_config.num_layers,
    dropout=gru_config.dropout,
).to(device)
criterion = nn.MSELoss()
optimizer = torch.optim.Adam(model.parameters(), lr=gru_config.lr)

history = []
best_state = None
best_val_metrics = None
best_epoch = -1
best_val_mae = float("inf")
patience_counter = 0

for epoch in range(1, gru_config.epochs + 1):
    model.train()
    train_losses = []

    for xb, yb in train_loader:
        xb = xb.to(device)
        yb = yb.to(device)
        optimizer.zero_grad()
        preds = model(xb)
        loss = criterion(preds, yb)
        loss.backward()
        optimizer.step()
        train_losses.append(loss.item())

    train_loss = float(np.mean(train_losses))
    val_metrics, _, _ = evaluate_model(model, val_loader)
    history.append(
        {
            "epoch": epoch,
            "train_loss": train_loss,
            "val_mae": val_metrics["mae"],
            "val_rmse": val_metrics["rmse"],
            "val_mape": val_metrics["mape"],
        }
    )

    print(
        f"Epoch {epoch:02d} | train_loss={train_loss:.4f} | "
        f"val_mae={val_metrics['mae']:.4f} | val_rmse={val_metrics['rmse']:.4f}"
    )

    if val_metrics["mae"] < best_val_mae:
        best_val_mae = val_metrics["mae"]
        best_val_metrics = val_metrics
        best_epoch = epoch
        best_state = {k: v.cpu().clone() for k, v in model.state_dict().items()}
        patience_counter = 0
    else:
        patience_counter += 1

    if patience_counter >= gru_config.patience:
        print("Early stopping triggered.")
        break

history_df = pd.DataFrame(history)
display(history_df.tail())

model.load_state_dict(best_state)
torch.save(best_state, BEST_GRU_WEIGHTS_PATH)
BEST_GRU_CONFIG_PATH.write_text(
    json.dumps(
        {
            **asdict(gru_config),
            "seed": SEED,
            "device_used": str(device),
            "best_epoch": best_epoch,
            "split_summary": split_summary,
            "scaler": "StandardScaler(train target only)",
        },
        indent=2,
        ensure_ascii=False,
    ),
    encoding="utf-8",
)

fig, axes = plt.subplots(1, 2, figsize=(14, 4))
axes[0].plot(history_df["epoch"], history_df["train_loss"], label="train_loss", color="tab:blue")
axes[0].set_title("GRU train loss")
axes[0].legend()
axes[1].plot(history_df["epoch"], history_df["val_mae"], label="val_mae", color="tab:orange")
axes[1].plot(history_df["epoch"], history_df["val_rmse"], label="val_rmse", color="tab:green")
axes[1].set_title("GRU validation metrics")
axes[1].legend()
plt.tight_layout()
plt.savefig(FIGURES_DIR / "gru_learning_curves.png", dpi=150, bbox_inches="tight")
plt.show()

r1_run = make_run_record(
    "R1",
    f"GRU(hidden={gru_config.hidden_size}, layers={gru_config.num_layers})",
    f"target windows only, window_size={gru_config.window_size}",
    gru_config.window_size,
    "StandardScaler(train target only)",
    "Adam",
    gru_config.lr,
    int(history_df["epoch"].max()),
    best_val_metrics,
    split_summary,
    notes=f"Best epoch: {best_epoch}",
)

all_runs_df = pd.concat([baseline_results_df, pd.DataFrame([r1_run])], ignore_index=True)
display(
    all_runs_df[
        ["experiment_id", "best_val_mae", "best_val_rmse", "best_val_mape", "model_summary"]
    ].sort_values("best_val_mae")
)

fig, ax = plt.subplots(figsize=(10, 5))
compare_df = all_runs_df.sort_values("best_val_mae").reset_index(drop=True)
ax.bar(
    compare_df["experiment_id"],
    compare_df["best_val_mae"],
    color=["#59A14F", "#E15759", "#4E79A7", "#F28E2B"],
)
ax.set_title("B1 vs B2 vs B3 vs R1 by validation MAE")
ax.set_ylabel("Validation MAE")
plt.tight_layout()
plt.savefig(FIGURES_DIR / "baselines_compare.png", dpi=150, bbox_inches="tight")
plt.show()

# Финальная оценка только для лучшей модели по validation.
best_experiment_id = all_runs_df.sort_values("best_val_mae").iloc[0]["experiment_id"]
test_truth = baseline_test_df["target"].to_numpy()
test_metrics = None
plot_pred = None
plot_dates = None

if best_experiment_id == "B1":
    plot_pred = baseline_test_df["lag_1"].to_numpy()
    test_metrics = collect_metrics(test_truth, plot_pred)
    plot_dates = baseline_test_df["date"].to_numpy()
elif best_experiment_id == "B2":
    plot_pred = baseline_test_df["rolling_mean_24"].to_numpy()
    test_metrics = collect_metrics(test_truth, plot_pred)
    plot_dates = baseline_test_df["date"].to_numpy()
elif best_experiment_id == "B3":
    X_test = x_scaler.transform(baseline_test_df[feature_columns])
    plot_pred = ridge.predict(X_test)
    test_metrics = collect_metrics(test_truth, plot_pred)
    plot_dates = baseline_test_df["date"].to_numpy()
else:
    test_metrics, test_truth, plot_pred = evaluate_model(model, test_loader)
    plot_dates = df["date"].iloc[val_end:].to_numpy()

print("Best experiment by validation MAE:", best_experiment_id)
print("Final test metrics:")
print(json.dumps(test_metrics, indent=2))

all_runs_df["test_mae"] = np.nan
all_runs_df["test_rmse"] = np.nan
all_runs_df["test_mape"] = np.nan
for metric_name, metric_value in test_metrics.items():
    all_runs_df.loc[all_runs_df["experiment_id"] == best_experiment_id, f"test_{metric_name}"] = metric_value
all_runs_df.to_csv(RUNS_PATH, index=False)
display(all_runs_df.sort_values("best_val_mae"))

fig, ax = plt.subplots(figsize=(15, 5))
ax.plot(plot_dates, test_truth, label="actual", color="tab:blue")
ax.plot(plot_dates, plot_pred, label=f"prediction ({best_experiment_id})", color="tab:red", alpha=0.8)
ax.set_title("Best model forecast on test")
ax.legend()
plt.tight_layout()
plt.savefig(FIGURES_DIR / "best_forecast_test.png", dpi=150, bbox_inches="tight")
plt.show()

print("Saved files:")
for path in [
    RUNS_PATH,
    BEST_GRU_WEIGHTS_PATH,
    BEST_GRU_CONFIG_PATH,
    FIGURES_DIR / "series_split.png",
    FIGURES_DIR / "baselines_compare.png",
    FIGURES_DIR / "gru_learning_curves.png",
    FIGURES_DIR / "best_forecast_test.png",
]:
    print("-", path.name, path.exists())
```
